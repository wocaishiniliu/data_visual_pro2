```python
plt.rcParams['font.sans-serif']=['SimHei'] 
plt.rcParams['axes.unicode_minus']=False
matplotlib.rcParams['axes.unicode_minus']=False
fig,ax=plt.subplots(2,2,figsize=(20,12))
plt.rcParams['font.sans-serif'] = ['SimHei']
a1 = pd.read_csv(r".\A曲线.csv").sort_values(by = ["x"])
a2 = pd.read_csv(r".\A虚线.csv").sort_values(by = ["x"])
b1 = pd.read_csv(r".\B曲线.csv").sort_values(by = ["x"])
b2 = pd.read_csv(r".\B虚线.csv").sort_values(by = ["x"])
c_blue = pd.read_csv(r".\C蓝.csv").sort_values(by = ["x"])
c_yellow = pd.read_csv(r".\C黄.csv").sort_values(by = ["x"])
c_red = pd.read_csv(r".\C红.csv").sort_values(by = ["x"])
c_green = pd.read_csv(r".\C绿.csv").sort_values(by = ["x"])
d_blue = pd.read_csv(r".\D蓝.csv").sort_values(by = ["x"])
d_yellow = pd.read_csv(r".\D黄.csv").sort_values(by = ["x"])
d_red = pd.read_csv(r".\D红.csv").sort_values(by = ["x"])
d_green = pd.read_csv(r".\D绿.csv").sort_values(by = ["x"])
ax[0,0].add_patch(patches.Rectangle((19, 0),width=73, height=100, color='#E7E1E1'))
ax[0,0].plot(a1["x"], a1["y"],color='darkorange',linewidth=5)
ax[0,0].plot(a2["x"], a2["y"],color='tan',linewidth=3,linestyle='--')
ax[0,0].set_yticks([0,25,50,75,100])
ax[0,0].set_xticks([19,28,58,89])
ax[0,0].set_xticklabels(['Jan23','Feb 1','Mar 1','Apr 1'],fontdict={'size':15})
ax[0,0].set_ylim(0, 100)
ax[0,0].set_ylabel('Emmissions(kt NOx/day)A',fontdict={'size':18},labelpad=2)
ax[0,0].set_xlabel('Data',fontdict={'size':18},labelpad=2)
ax[0,0].text(0,90.55,"NOx 2020",fontsize=15)
ax[0,0].annotate(text='Bottom-up estimate',xy=(17,53),xytext=(25,55),arrowprops=dict(arrowstyle = '-'))
ax[0,0].annotate(text='TROPOMI-constrained',xy=(35,33.5),xytext=(33,25),arrowprops=dict(arrowstyle = '-'))
ax[0,1].add_patch(patches.Rectangle((19, 0),width=73, height=100, color='#E7E1E1'))
ax[0,1].set_ylim(0, 60)
ax[0,1].plot(b1["x"], b1["y"],color='red',linewidth=5)
ax[0,1].plot(b2["x"], b2["y"], dashes=[6,2],color='tan',linewidth=3,linestyle='--')
ax[0,1].set_yticks([0,15,30,45,60])
ax[0,1].set_ylabel('Emmissions(Mt CO2/day)B',fontdict={'size':18},labelpad=2)
ax[0,1].text(0,55.55,"CO2 2020",fontsize=15)
ax[0,1].annotate(text='Bottom-up estimate',xy=(17,28),xytext=(25,33),arrowprops=dict(arrowstyle = '-'))
ax[0,1].annotate(text='TROPOMI-constrained',xy=(34,19),xytext=(33,13),arrowprops=dict(arrowstyle = '-'))
ax[0,1].set_xlabel('Data',fontdict={'size':18},labelpad=2)
ax[0,1].set_xticks([19,28,58,89])
ax[0,1].set_xticklabels(['Jan23','Feb 1','Mar 1','Apr 1'],fontdict={'size':15})
ax[1,0].add_patch(patches.Rectangle((19, -6),width=73, height=12, color='#E7E1E1'))
ax[1,0].set_ylim(-6, 6)
ax[1,0].plot(c_blue["x"], c_blue["y"],c="blue",linewidth=3)
ax[1,0].plot(c_yellow["x"], c_yellow["y"], c="yellow",linewidth=3)
ax[1,0].plot(c_red["x"], c_red["y"], c="red",linewidth=3)
ax[1,0].plot(c_green["x"], c_green["y"], c="green",linewidth=3)
ax[1,0].set_yticks([-6,-3,0,3,6])
ax[1,0].set_ylabel('▲Emmissions(Mt NOx/day)B',fontdict={'size':18},labelpad=2)
ax[1,0].text(0,4.55,"▲NOx 2020 = TROPOMI - bottom-up",fontsize=15)
ax[1,0].annotate(text='Residential',xy=(22,0.5),xytext=(18,2.3),arrowprops=dict(arrowstyle = '-'))
ax[1,0].annotate(text='Power',xy=(62,0.65),xytext=(60,1.2),arrowprops=dict(arrowstyle = '-'))
ax[1,0].annotate(text='Industry',xy=(25,-2.8),xytext=(27,-4),arrowprops=dict(arrowstyle = '-'))
ax[1,0].annotate(text='Transport',xy=(77,-2.3),xytext=(69,-3.5),arrowprops=dict(arrowstyle = '-'))
ax[1,0].set_xlabel('Data',fontdict={'size':18},labelpad=2)
ax[1,0].set_xticks([19,28,58,89])
ax[1,0].set_xticklabels(['Jan23','Feb 1','Mar 1','Apr 1'],fontdict={'size':15})
ax[1,1].add_patch(patches.Rectangle((19, -6),width=73, height=12, color='#E7E1E1'))
ax[1,1].set_ylim(-6, 6)
ax[1,1].plot(d_blue["x"], d_blue["y"],c="blue",linewidth=3)
ax[1,1].plot(d_yellow["x"], d_yellow["y"], c="yellow",linewidth=3)
ax[1,1].plot(d_red["x"], d_red["y"], c="red",linewidth=3)
ax[1,1].plot(d_green["x"], d_green["y"], c="green",linewidth=3)
ax[1,1].set_yticks([-6,-3,0,3,6])
ax[1,1].set_ylabel('▲Emmissions(Mt CO2/day)B',fontdict={'size':18},labelpad=2)
ax[1,1].text(0,4.55,"▲CO2 2020 = TROPOMI - bottom-up",fontsize=15)
ax[1,1].annotate(text='Residential',xy=(21,1.2),xytext=(18,2.3),arrowprops=dict(arrowstyle = '-'))
ax[1,1].annotate(text='Power',xy=(72,1.5),xytext=(69,2.5),arrowprops=dict(arrowstyle = '-'))
ax[1,1].annotate(text='Industry',xy=(30,-2.8),xytext=(27,-4),arrowprops=dict(arrowstyle = '-'))
ax[1,1].annotate(text='Transport',xy=(82,-1),xytext=(79,-2),arrowprops=dict(arrowstyle = '-'))
ax[1,1].set_xlabel('Data',fontdict={'size':18},labelpad=2)
ax[1,1].set_xticks([19,28,58,89])
ax[1,1].set_xticklabels(['Jan23','Feb 1','Mar 1','Apr 1'],fontdict={'size':15})
plt.show()
```
